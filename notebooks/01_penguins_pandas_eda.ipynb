## 3. Indexing과 Slicing

## 4. Filtering (조건 필터링)

조건식을 대괄호 안에 넣어 조건을 만족하는 행만 추출합니다.
``` python
df[ df['col'] > 100 ]                   # 단일 조건
df[ (df['a'] > 1) & (df['b'] < 5) ]     # AND 조건
df[ (df['a'] > 1) | (df['b'] < 5) ]     # OR 조건
df[ df['col'].isin(['A', 'B']) ]        # 여러 값 중 하나
```

각 조건은 괄호로 묶어야 하며, AND는 `&`, OR는 `|`를 사용합니다.
